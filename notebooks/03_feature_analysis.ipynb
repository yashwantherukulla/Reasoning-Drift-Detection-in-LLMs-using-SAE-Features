# Phase 4: Feature Analysis Exploration

Interactive exploration of Reasoning Drift Score (RDS) and other metrics across problems and layers.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path

METRICS_DETAIL = Path("../results/metrics/metrics_detail.csv")
STABILITY_DETAIL = Path("../results/metrics/stability_detail.csv")

if METRICS_DETAIL.exists():
    detail = pd.read_csv(METRICS_DETAIL)
    detail['category'] = detail['problem_id'].str.extract(r'^([a-z]+)').replace({'gsm': 'gsm8k'})
    display(detail.head())
else:
    print(f"File {METRICS_DETAIL} not found. Run Phase 4 analysis first.")
```

## RDS Across Layers (Interactive)

```python
if 'detail' in locals():
    fig = px.line(detail, x="layer", y="rds", color="problem_id", line_dash="cond_b", 
                  facet_col="cond_a", title="RDS Across Layers per Problem",
                  hover_data=["category", "cosine_sim", "drift_dir_cosine"])
    fig.show()
```

## Drift Direction Cosine vs RDS

```python
if 'detail' in locals():
    fig = px.scatter(detail, x="rds", y="drift_dir_cosine", color="category", 
                     hover_name="problem_id", facet_col="layer",
                     title="Drift Direction vs RDS by Layer")
    fig.show()
```
