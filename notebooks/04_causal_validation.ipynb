# Phase 5: Causal Validation (Patching)

Interactive exploration of causal patching results.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path

PATCHING_RESULTS = Path("../results/patching/patching_results.csv")

if PATCHING_RESULTS.exists():
    patching = pd.read_csv(PATCHING_RESULTS)
    display(patching.head())
else:
    print(f"File {PATCHING_RESULTS} not found. Run Phase 5 patching first.")
```

## % Answer Changed by Features Patched & Layer

```python
if 'patching' in locals() and not patching.empty:
    grp = patching.groupby(["layer_patched", "n_features_patched"])["answer_changed"].mean().reset_index()
    fig = px.bar(grp, x="layer_patched", y="answer_changed", color="n_features_patched", barmode="group",
                 title="Fraction of Answers Changed by Patching")
    fig.show()
else:
    print("Not enough data to plot.")
```
